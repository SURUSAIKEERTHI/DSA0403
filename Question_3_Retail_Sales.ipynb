Problem Statement

A supermarket wants to analyze product sales based on category, price, quantity sold, discount, and revenue. Exploratory Data Analysis (EDA) is performed to identify high-selling products, understand revenue patterns, analyze relationships between sales variables, and support better business decisions.

Purpose and Significance

The purpose of this dataset is to analyze product sales performance across different categories. The analysis helps identify high-selling products, high-revenue categories, pricing patterns, and relationships between price, quantity sold, discount, and revenue.

This analysis is significant because it can help supermarket management improve inventory planning, pricing strategies, promotional campaigns, and product stocking decisions.

Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2: Create the Dataset

```python
data = {
    "Product_ID": [
        "PR01","PR02","PR03","PR04","PR05",
        "PR06","PR07","PR08","PR09","PR10",
        "PR11","PR12","PR13","PR14","PR15",
        "PR16","PR17","PR18","PR19","PR20"
    ],

    "Category": [
        "Grocery","Snacks","Dairy","Fruits","Vegetables",
        "Grocery","Snacks","Dairy","Fruits","Vegetables",
        "Grocery","Snacks","Dairy","Fruits","Vegetables",
        "Grocery","Snacks","Dairy","Fruits","Vegetables"
    ],

    "Price": [
        50,30,45,60,40,
        80,25,55,70,35,
        90,20,65,75,45,
        100,35,50,80,30
    ],

    "Quantity_Sold": [
        20,35,18,25,30,
        15,40,16,22,28,
        12,45,14,20,26,
        10,32,19,18,35
    ],

    "Discount": [
        5,3,2,5,4,
        6,3,2,5,4,
        6,2,3,5,4,
        7,3,2,6,3
    ],

    "Revenue": [
        1000,1050,810,1500,1200,
        1200,1000,880,1540,980,
        1080,900,910,1500,1170,
        1000,1120,950,1440,1050
    ]
}

df = pd.DataFrame(data)

print(df)
```

Step 3: Dataset Description

```python
print("Shape of Dataset:", df.shape)
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nColumn Names")
print(df.columns)

print("\nData Types")
print(df.dtypes)

print("\nDataset Information")
df.info()
```

The dataset contains 20 product records and 6 features.

Product_ID – Unique product

identifier – Categorical

Category – Product category – Categorical

Price – Price of the product – Numerical

Quantity_Sold – Number of units sold – Numerical

Discount – Discount offered – Numerical

Revenue – Revenue generated – Numerical

Step 4: Check and Handle Missing Values

```python
print("Missing Values")
print(df.isnull().sum())

df = df.dropna()

print("\nMissing Values After Preprocessing")
print(df.isnull().sum())
```

Step 5: Check and Remove Duplicates

```python
print("Duplicate Rows:", df.duplicated().sum())

df = df.drop_duplicates()

print("Shape After Removing Duplicates:", df.shape)
```

Step 6: Check Inconsistencies and Preprocessing

```python
print("Unique Product Categories:")
print(df["Category"].unique())

print("\nPrice Range:")
print(df["Price"].min(), "to", df["Price"].max())

print("\nQuantity Sold Range:")
print(df["Quantity_Sold"].min(), "to", df["Quantity_Sold"].max())

print("\nDiscount Range:")
print(df["Discount"].min(), "to", df["Discount"].max())

print("\nRevenue Range:")
print(df["Revenue"].min(), "to", df["Revenue"].max())
```

Step 7: Descriptive Statistics

```python
print(df.describe())
```

```python
print("Mean")
print(df.mean(numeric_only=True))

print("\nMedian")
print(df.median(numeric_only=True))

print("\nMinimum")
print(df.min(numeric_only=True))

print("\nMaximum")
print(df.max(numeric_only=True))

print("\nStandard Deviation")
print(df.std(numeric_only=True))
```

Step 8: Numerical Variable Analysis

```python
numerical_columns = [
    "Price",
    "Quantity_Sold",
    "Discount",
    "Revenue"
]

print(df[numerical_columns].describe())
```

Step 9: Categorical Variable Analysis

```python
print("Product Category Distribution")
print(df["Category"].value_counts())
```

Step 10: Category-Wise Sales Analysis

```python
category_analysis = df.groupby("Category")[
    ["Price", "Quantity_Sold", "Discount", "Revenue"]
].mean()

print(category_analysis)
```

Step 11: Correlation Analysis

```python
correlation = df.corr(numeric_only=True)

print(correlation)
```

Step 12: Histogram – Revenue Distribution

```python
plt.figure(figsize=(7,5))

plt.hist(
    df["Revenue"],
    bins=6,
    edgecolor="black"
)

plt.title("Distribution of Product Revenue")
plt.xlabel("Revenue")
plt.ylabel("Number of Products")

plt.show()
```

Step 13: Bar Chart – Total Revenue by Category

```python
category_revenue = df.groupby("Category")["Revenue"].sum().sort_values(
    ascending=False
)

plt.figure(figsize=(8,5))

category_revenue.plot(kind="bar")

plt.title("Total Revenue by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)

plt.show()
```

Step 14: Scatter Plot – Price vs Revenue

```python
plt.figure(figsize=(7,5))

sns.scatterplot(
    data=df,
    x="Price",
    y="Revenue",
    hue="Category",
    s=100
)

plt.title("Price vs Revenue by Product Category")
plt.xlabel("Product Price")
plt.ylabel("Revenue")
plt.legend(title="Category")

plt.show()
```

Step 15: Box Plot – Revenue by Category

```python
plt.figure(figsize=(9,5))

sns.boxplot(
    data=df,
    x="Category",
    y="Revenue"
)

plt.title("Revenue Distribution by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Revenue")

plt.show()
```

Step 16: Correlation Heatmap

```python
plt.figure(figsize=(7,5))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap of Retail Sales Variables")

plt.show()
```

```python
high_selling = df.sort_values(
    by="Quantity_Sold",
    ascending=False
).head(5)

print("Top 5 High-Selling Products")
print(high_selling[
    ["Product_ID", "Category", "Quantity_Sold"]
])
```
