Problem Statement

A hospital wants to analyze patient health risk based on age, blood sugar, blood pressure, BMI, and disease status. EDA is performed to identify important health patterns and factors associated with disease status.

Purpose and Significance

The purpose of this dataset is to analyze the relationship between patient health indicators and disease status. The analysis can help identify patterns associated with higher health risk and support healthcare professionals in early screening and data-driven decision-making.

Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2: Create the Dataset

```python
data = {
    "Patient_ID": [
        "P01","P02","P03","P04","P05","P06","P07","P08","P09","P10",
        "P11","P12","P13","P14","P15","P16","P17","P18","P19","P20"
    ],

    "Age": [
        45,32,55,28,60,40,52,35,48,30,
        62,38,50,27,58,42,46,34,53,29
    ],

    "Sugar_Level": [
        145,98,180,90,200,120,165,105,155,95,
        210,115,170,88,190,130,150,100,175,92
    ],

    "BP": [
        130,118,145,110,150,125,140,120,135,115,
        155,122,142,108,148,128,132,116,144,112
    ],

    "BMI": [
        28,23,31,22,34,26,30,24,29,23,
        35,25,30,21,33,27,28,24,31,22
    ],

    "Disease_Status": [
        "Yes","No","Yes","No","Yes","No","Yes","No","Yes","No",
        "Yes","No","Yes","No","Yes","No","Yes","No","Yes","No"
    ]
}

df = pd.DataFrame(data)

print(df)
```

Step 3: Dataset Description

```python
print("Shape of Dataset:", df.shape)
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nColumn Names")
print(df.columns)

print("\nData Types")
print(df.dtypes)

print("\nDataset Information")
df.info()
```

Step 4: Check and Handle Missing Values

```python
print("Missing Values")
print(df.isnull().sum())

df = df.dropna()

print("\nMissing Values After Preprocessing")
print(df.isnull().sum())
```

Step 5: Check and Remove Duplicates

```python
print("Duplicate Rows:", df.duplicated().sum())

df = df.drop_duplicates()

print("Shape After Removing Duplicates:", df.shape)
```

Step 6: Check Inconsistencies and Preprocessing

```python
print("Unique Disease Status Values:")
print(df["Disease_Status"].unique())

print("\nAge Range:")
print(df["Age"].min(), "to", df["Age"].max())

print("\nSugar Level Range:")
print(df["Sugar_Level"].min(), "to", df["Sugar_Level"].max())

print("\nBP Range:")
print(df["BP"].min(), "to", df["BP"].max())

print("\nBMI Range:")
print(df["BMI"].min(), "to", df["BMI"].max())
```

Step 7: Descriptive Statistics

```python
print(df.describe())
```

```python
print("Mean")
print(df.mean(numeric_only=True))

print("\nMedian")
print(df.median(numeric_only=True))

print("\nMinimum")
print(df.min(numeric_only=True))

print("\nMaximum")
print(df.max(numeric_only=True))

print("\nStandard Deviation")
print(df.std(numeric_only=True))
```

Step 8: Numerical Variable Analysis

```python
numerical_columns = ["Age", "Sugar_Level", "BP", "BMI"]

print(df[numerical_columns].describe())
```

Step 9: Categorical Variable Analysis

```python
print("Disease Status Distribution")
print(df["Disease_Status"].value_counts())
```

Step 10: Health Indicators Based on Disease Status

```python
disease_analysis = df.groupby("Disease_Status")[
    ["Age", "Sugar_Level", "BP", "BMI"]
].mean()

print(disease_analysis)
```

Step 11: Correlation Analysis

```python
correlation = df.corr(numeric_only=True)

print(correlation)
```

Step 12: Histogram – Sugar Level Distribution

```python
plt.figure(figsize=(7,5))

plt.hist(
    df["Sugar_Level"],
    bins=6,
    edgecolor="black"
)

plt.title("Distribution of Patient Sugar Levels")
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")

plt.show()
```

Step 13: Bar Chart – Disease Status

```python
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="Disease_Status"
)

plt.title("Distribution of Disease Status")
plt.xlabel("Disease Status")
plt.ylabel("Number of Patients")

plt.show()
```

Step 14: Scatter Plot – Sugar Level vs BMI

```python
plt.figure(figsize=(7,5))

sns.scatterplot(
    data=df,
    x="Sugar_Level",
    y="BMI",
    hue="Disease_Status",
    s=100
)

plt.title("Sugar Level vs BMI by Disease Status")
plt.xlabel("Sugar Level")
plt.ylabel("BMI")
plt.legend(title="Disease Status")

plt.show()
```

Step 15: Box Plot – Sugar Level by Disease Status

```python
plt.figure(figsize=(7,5))

sns.boxplot(
    data=df,
    x="Disease_Status",
    y="Sugar_Level"
)

plt.title("Sugar Level Distribution by Disease Status")
plt.xlabel("Disease Status")
plt.ylabel("Sugar Level")

plt.show()
```

Step 16: Correlation Heatmap

```python
plt.figure(figsize=(7,5))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap of Patient Health Indicators")

plt.show()
```

The healthcare dataset contains 20 patient records and 6 features.

Patients with disease generally show higher sugar levels, blood pressure, BMI, and age than patients without disease in this sample.
Higher sugar levels and BMI appear to be associated with disease-positive cases.
The numerical health indicators show relationships with one another, suggesting that multiple health factors should be considered together when analyzing patient risk.

These findings describe patterns in the provided dataset and should not be treated as medical diagnosis or proof of causation.

Conclusion
