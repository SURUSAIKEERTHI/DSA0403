Problem Statement:
A bank wants to analyze loan approval decisions based on income, credit score, loan amount, employment type, and loan approval status. Exploratory Data Analysis (EDA) is performed to identify important patterns associated with loan approval and rejection.
Purpose and Significance:
The purpose of this dataset is to analyze the factors that may influence loan approval decisions, such as customer income, credit score, requested loan amount, and employment type.
This analysis is significant because it helps banks understand patterns in historical loan decisions, identify characteristics associated with approved and rejected applications, and support more consistent and data-driven loan assessment processes.

Step 1: Import Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2: Create the Dataset

```python
data = {
    "Customer_ID": [
        "C01","C02","C03","C04","C05",
        "C06","C07","C08","C09","C10",
        "C11","C12","C13","C14","C15",
        "C16","C17","C18","C19","C20"
    ],

    "Income": [
        45000,25000,60000,22000,50000,
        30000,70000,28000,55000,24000,
        65000,35000,48000,26000,75000,
        32000,52000,27000,68000,23000
    ],

    "Credit_Score": [
        720,580,750,560,710,
        600,780,590,730,570,
        760,620,700,585,790,
        610,725,575,770,555
    ],

    "Loan_Amount": [
        200000,150000,300000,120000,250000,
        180000,350000,160000,270000,130000,
        320000,200000,240000,140000,400000,
        190000,260000,150000,330000,125000
    ],

    "Employment_Type": [
        "Salaried","Self-employed","Salaried","Unemployed","Salaried",
        "Self-employed","Salaried","Self-employed","Salaried","Unemployed",
        "Salaried","Self-employed","Salaried","Unemployed","Salaried",
        "Self-employed","Salaried","Self-employed","Salaried","Unemployed"
    ],

    "Loan_Status": [
        "Approved","Rejected","Approved","Rejected","Approved",
        "Rejected","Approved","Rejected","Approved","Rejected",
        "Approved","Rejected","Approved","Rejected","Approved",
        "Rejected","Approved","Rejected","Approved","Rejected"
    ]
}

df = pd.DataFrame(data)

print(df)
```

Step 3: Dataset Description

```python
print("Shape of Dataset:", df.shape)
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nColumn Names")
print(df.columns)

print("\nData Types")
print(df.dtypes)

print("\nDataset Information")
df.info()
```

The dataset contains 20 customer records and 6 features.

Customer_ID – Unique identification of each customer – Categorical
Income – Customer's income – Numerical
Credit_Score – Customer's credit score – Numerical
Loan_Amount – Amount of loan requested – Numerical
Employment_Type – Employment status of the customer – Categorical
Loan_Status – Loan decision (Approved/Rejected) – Categorical

Step 4: Check and Handle Missing Values

```python
print("Missing Values")
print(df.isnull().sum())

df = df.dropna()

print("\nMissing Values After Preprocessing")
print(df.isnull().sum())
```

Step 5: Check and Remove Duplicates

```python
print("Duplicate Rows:", df.duplicated().sum())

df = df.drop_duplicates()

print("Shape After Removing Duplicates:", df.shape)
```

Step 6: Check Inconsistencies and Preprocessing

```python
print("Unique Employment Types:")
print(df["Employment_Type"].unique())

print("\nUnique Loan Status Values:")
print(df["Loan_Status"].unique())

print("\nIncome Range:")
print(df["Income"].min(), "to", df["Income"].max())

print("\nCredit Score Range:")
print(df["Credit_Score"].min(), "to", df["Credit_Score"].max())

print("\nLoan Amount Range:")
print(df["Loan_Amount"].min(), "to", df["Loan_Amount"].max())
```

Step 7: Descriptive Statistics

```python
print(df.describe())
```

```python
print("Mean")
print(df.mean(numeric_only=True))

print("\nMedian")
print(df.median(numeric_only=True))

print("\nMinimum")
print(df.min(numeric_only=True))

print("\nMaximum")
print(df.max(numeric_only=True))

print("\nStandard Deviation")
print(df.std(numeric_only=True))
```

Step 8: Numerical Variable Analysis

```python
numerical_columns = [
    "Income",
    "Credit_Score",
    "Loan_Amount"
]

print(df[numerical_columns].describe())
```

Step 9: Categorical Variable Analysis

```python
print("Employment Type Distribution")
print(df["Employment_Type"].value_counts())

print("\nLoan Status Distribution")
print(df["Loan_Status"].value_counts())
```

Step 10: Average Financial Factors by Loan Status

```python
loan_analysis = df.groupby("Loan_Status")[
    ["Income", "Credit_Score", "Loan_Amount"]
].mean()

print(loan_analysis)
```

Step 11: Correlation Analysis

```python
correlation = df.corr(numeric_only=True)

print(correlation)
```

Step 12: Histogram – Credit Score Distribution

```python
plt.figure(figsize=(7,5))

plt.hist(
    df["Credit_Score"],
    bins=6,
    edgecolor="black"
)

plt.title("Distribution of Customer Credit Scores")
plt.xlabel("Credit Score")
plt.ylabel("Number of Customers")

plt.show()
```

Step 13: Bar Chart – Loan Approval Status

```python
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="Loan_Status"
)

plt.title("Distribution of Loan Approval Status")
plt.xlabel("Loan Status")
plt.ylabel("Number of Customers")

plt.show()
```

Step 14: Scatter Plot – Income vs Credit Score

```python
plt.figure(figsize=(7,5))

sns.scatterplot(
    data=df,
    x="Income",
    y="Credit_Score",
    hue="Loan_Status",
    s=100
)

plt.title("Income vs Credit Score by Loan Status")
plt.xlabel("Customer Income")
plt.ylabel("Credit Score")
plt.legend(title="Loan Status")

plt.show()
```

Step 15: Box Plot – Credit Score by Loan Status

```python
plt.figure(figsize=(7,5))

sns.boxplot(
    data=df,
    x="Loan_Status",
    y="Credit_Score"
)

plt.title("Credit Score Distribution by Loan Status")
plt.xlabel("Loan Status")
plt.ylabel("Credit Score")

plt.show()
```

Step 16: Correlation Heatmap

```python
plt.figure(figsize=(7,5))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap of Banking Loan Variables")

plt.show()
```
